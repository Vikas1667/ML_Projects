```python
from lxml import etree
import urllib.request
from html.parser import HTMLParser
from bs4 import BeautifulSoup
import requests

url="http://diehardindian.com/ward-offices-and-corporators/"
# web = urllib.request.urlopen(url)

s = web.read().decode('utf-8')
# print(s)

html_content=requests.get(url).text
print(html_content)
soup = BeautifulSoup(html_content, "lxml")
print(soup.prettify())
```

```python

table = soup.find_all("table")[2].children
print(table)
rows = []

for child in soup.find_all("table")[2].children:
    print(child)
    row = []

    for td in child:
        try:
            text=td.text
            print(text)
            
            row.append(td.text.replace('\n', '|'))
        except:
            continue
    if len(row) > 0:
        rows.append(row)

```

```python
print(rows)
```

```python
columns_values=rows[0][0].lstrip('|').rstrip('|').split('|')
print(columns_values)
rows_values=[i.lstrip('|').rstrip('|').split('|') for i in rows[1:][0]]
rows_values=[['']+i if len(i)!=5 else i for i in rows_values]
print(rows_values)
```

```python
path = r'./Artisan_data_analysis/' # use your path
all_files = glob.glob(path + "/*.xlsx")
print(all_files)
li = []

for filename in all_files:
    df = pd.read_excel(filename, index_col=None, header=0)
    li.append(df)

frame = pd.concat(li, axis=0, ignore_index=True)

frame.isnull().sum()
```

```python
# display(frame)
m=frame['FATHER_AND_SPOUSES_NAME'].mode()
print(m[0])


frame['FATHER_AND_SPOUSES_NAME'].fillna(frame['FATHER_AND_SPOUSES_NAME'].mode()[0],inplace=True)
frame['Social_Group_SC_ST_OBC_GEN_MIN'].fillna(frame['Social_Group_SC_ST_OBC_GEN_MIN'].mode()[0],inplace=True)
frame['NAME_OF_CRAFT'].fillna(frame['NAME_OF_CRAFT'].mode()[0],inplace=True)
frame['ARTISIAN_NAME'].fillna(frame['ARTISIAN_NAME'].mode()[0],inplace=True)
frame['VILLAGE'].fillna(frame['VILLAGE'].mode()[0],inplace=True)
frame['TOWN'].fillna(frame['TOWN'].mode()[0],inplace=True)
frame['CITY'].fillna(frame['CITY'].mode()[0],inplace=True)
frame['DISTRICT'].fillna(frame['DISTRICT'].mode()[0],inplace=True)
frame['STATE'].fillna(frame['STATE'].mode()[0],inplace=True)
```

```python
# frame.isna().count()
frame.isnull().sum()
```
